## b. Finding and addressing mixed data types 

```python
#Finding mixed data column
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
#Fixing mixed type data
df['City']=df['City'].astype('str')
df['Wind_Direction']=df['Wind_Direction'].astype('str')
df['Weather_Condition']=df['Weather_Condition'].astype('str')
df['Sunrise_Sunset']=df['Sunrise_Sunset'].astype('str')
```

# 05. Data merging

```python
#Merging the dataset containing the number of drivers per state. First importing the additional dataset
df_drivers=pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Licensed_drivers_reformatted.csv'))
```

```python
#Ensuring that columns in both datasets have similar type
df.dtypes
```

```python
df_drivers.dtypes
```

```python
#Renaming year in df and reformatting year in df_drivers
df=df.rename(columns={'year':'Year'})
df['Year']=df['Year'].astype('int64')
```

```python
#Reformatting State in both datasets
df['State']=df['State'].astype('str')
df_drivers['State']=df_drivers['State'].astype('str')
```

```python
df_drivers.head(100)
```

```python
#Merging datasets
df_merged=df.merge(df_drivers,on=['State', 'Year'], how='left', indicator=True)
```
