```python
#Creating a subset of data for correlation matrix
sub =df[['Severity', 'Distance(mi)','Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Weather_Condition', 'Year', 'Month', 'Hour','Weather', 'Licensed_Drivers']]
```

```python
#Creating a correlation heatmap using matplotlib
corr1=plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(10, 10)) # figure size 
corr2=plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=12, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
plt.show()
```

```python
#Calculating IQR and replacing the outliers with missing values
#for x in ['City_accidents_per_day']:
    q75,q25 = np.percentile(merged1.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    merged1.loc[merged1[x] < min,x] = np.nan
    merged1.loc[merged1[x] > max,x] = np.nan
```
